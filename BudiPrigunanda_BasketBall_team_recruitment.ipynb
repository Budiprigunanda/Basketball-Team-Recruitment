### No. 3

```python
data_model = nba.drop(['gp','net_rating','oreb_pct','dreb_pct','ts_pct','draft_number','college','draft_year','draft_round','team_abbreviation','country','season','player_name'],axis=1)
X = data_model.drop(columns='potential_player')
y = data_model['potential_player']

X_train_val, X_test, y_train_val, y_test = train_test_split(X,y,test_size=.2, stratify=y, random_state=2020)
```

### dari syntax di atas, saya menghapus beberapa kolom yang ada di nba untuk membuat classification models

### No. 4

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.feature_selection import SelectPercentile, univariate_selection, RFE

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score, precision_score, plot_confusion_matrix
```

```python
data_model.info()
```

```python
rfc = RandomForestClassifier()
dtc = DecisionTreeClassifier()
```

#### saya menggunakan classification models random forest dan decision tree

```python
smote = SMOTE()
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV, cross_val_score, RepeatedStratifiedKFold,\
StratifiedKFold
```

```python
import sklearn
sklearn.metrics.SCORERS.keys()
```

```python
rfc_pipe = Pipeline([
    ('rfe', RFE(rfc)),
    ('resampling', smote),
    ('rfc', rfc)
])

dtc_pipe = Pipeline([
    ('rfe', RFE(dtc)),
    ('resampling', smote),
    ('dtc', dtc)
])
```

```python
rskfold = RepeatedStratifiedKFold(n_splits=5, random_state=2020)

rfc_score = cross_val_score(rfc_pipe, X_train_val, y_train_val, scoring='precision', cv=rskfold, n_jobs=-1, verbose=1)
dtc_score = cross_val_score(dtc_pipe, X_train_val, y_train_val, scoring='precision', cv=rskfold, n_jobs=-1, verbose=1)

```

```python
print('Random forest Score:' , rfc_score.mean())
print('Decision Tree Score:', dtc_score.mean())
```

#### dari hasil di atas, random forest memiliki nilai yang lebih tinggi dari decision tree

```python
rfc_pipe.get_params()
```

```python
rfc_params = {
    'rfc__max_depth': np.arange(8,15),
    'rfc__min_samples_split': np.arange(6,9)
}
```

```python
skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
rfc_grid = GridSearchCV(rfc_pipe, param_grid=rfc_params, scoring='precision', cv=skfold, n_jobs=-1, verbose=1)
rfc_grid.fit(X_train_val, y_train_val)
```

```python
rfc_tuned = rfc_grid.best_estimator_
rfc_tuned_score = cross_val_score(rfc_tuned, X_train_val, y_train_val, scoring='precision', cv=rskfold, n_jobs=-1, verbose=1)
rfc_tuned_score
```

```python
rfc_tuned_score.mean()
```

```python
rfc_tuned.fit(X_train_val, y_train_val)
plot_confusion_matrix(rfc_tuned, X_test, y_test)
```
